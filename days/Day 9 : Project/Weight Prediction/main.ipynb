```python
#import necessaries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("data.csv",index_col='Index')
print(data.head())
# Heights are in incs and weights are in pounds
```

```python
X = data.iloc[:,:-1].values
y = data.iloc[:,-1:].values
```

```python
print(X[:10])
```

```python
print(y[:10])
```

```python
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x= X, y= y)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X,y)
```

```python
y_pred = regression.predict(X)
```

```python
plt.xlabel('Height')
plt.ylabel('Weight')
plt.plot(X,y_pred)
```

```python
plt.scatter(x= X, y= y)
plt.plot(X,y_pred,color="red")
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title("Weight prediction by Height")
plt.savefig("result.jpg")
plt.show()
```
